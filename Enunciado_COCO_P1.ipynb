### Apartado 1: Tetracin
La tetracin es una hiperoperacin que se extiende ms all de la exponenciacin. El smbolo empleado para representar esta operacin es la doble flecha ascendente en la notacin Flecha de Knuth, como se muestra en la siguiente frmula:

\begin{equation}
a \uparrow\uparrow n = \begin{cases}
a & \text{si } n = 1 \\
a^{(a \uparrow\uparrow (n-1))} & \text{si } n > 1
\end{cases}
\end{equation}

Esto tambin se puede escribir usando la notacin de Rudy Rucker:

\begin{equation}
^{n}a = a^{a^{a^{\cdot^{\cdot^{\cdot^{a}}}}}} \quad \text{(}n \text{ veces)}
\end{equation}

En este apartado, se implementarn dos versiones de un algoritmo para calcular la tetracin: una versin recursiva y una versin iterativa.

Fuente recomendada: https://www.youtube.com/watch?v=r4L1cr2yPhE

Tetracin Recursiva

```python
""" Parmetros:
    a: int // Es la base de la tetracin
    b: int // Es el nmero de veces por el cual se desea elevar la base por si misma

    Variables:
    resultRec: int // Guarda el valor del resultado dentro de la funcin recursiva
    resultado: int // Guarda el valor del resultado final y es igual a el resultado de la llamada a la funcin recursiva

    Funciones:
    tetracionRecursiva: devuelve un int (que no tiene lmite como en C) y se encarga de realizar la tetracin
    de manera recursiva
"""

def tetracionRecursiva(a,b): #funcin recursiva que implementa la tetracin (similar a las vistas en clase)

    if b == 1:  # Caso especfico donde si "b" vale 1 el return es "a" por las propiedades de las potencias
        return a
    
    if b == 0:  # Caso especfico donde si "b" vale 0 el return es 1 por las propiedades de las potencias
        return 1
    
    resultRec = a ** tetracionRecursiva(a, b - 1)  #Llamada a la funcin recursiva. Se eleva b-1 veces como indica el enunciado

    print(f"en el nivel {b}, el resultado es : {resultRec}")

    return resultRec

# Bloque try-except (como try-catch en otros lenguajes) para manejar errores del usuario al ingresar los datos por teclado
try:
    a = int(input("Introduce el nmero al que le deseas realizar la tetracin: ")) # Hacemos uso de la funcin input() para que el usuario pueda ingresar los datos por teclado. Hay que declarar el tipo de dato de la entrada por teclado por que la funcin input() devuelve un string
    b = int(input("Introduce el nmero de veces que deseas elevar ese nmero por si mismo: "))

    resultado = tetracionRecursiva(a, b)

    print(f"El resultado de realizar la tetracin {b} veces a {a} es de: {resultado}")

except ValueError:   # Mensaje de la excepcin si el nmero no es entero
    print("Error al ingresar los datos. Por favor ingrese un nmero entero.")
```

Tetracin iterativa

```python
""" Parmetros:
    a: int // Es la base de la tetracin
    b: int // Es el nmero de veces por el cual se desea elevar la base por si misma

    Variables:
    result: int // Guarda el valor del resultado dentro de la funcin 
    resultado: int // Guarda el valor del resultado final y es igual a el resultado de la llamada a la funcin 

    Funciones:
    tetracionIterativa: devuelve un int (que no tiene lmite como en C) y se encarga de realizar la tetracin
    de manera iterativa 
"""

def tetracionIterativa(a,b):

    if b == 1:
        return a

    if b == 0:
        return 1

    resultItvo = a

    for i in range(1, b):
        resultItvo = a ** resultItvo
        print(f"En la iteracin {i}, el resultado parcial es: {resultItvo}")
    
    return resultItvo

try:
    a = int(input("Introduce el nmero al que le deseas realizar la tetracin: ")) # Hacemos uso de la funcin input() para que el usuario pueda ingresar los datos por teclado. Hay que declarar el tipo de dato de la entrada por teclado por que la funcin input() devuelve un string
    b = int(input("Introduce el nmero de veces que deseas elevar ese nmero por si mismo: "))

    resultado = tetracionIterativa(a, b)
    print(f"El resultado de realizar la tetracin {b} veces a {a} es de: {resultado}")

except ValueError:   # Mensaje de la excepcin si el nmero no es entero
    print("Error al ingresar los datos. Por favor ingrese un nmero entero.")
```
