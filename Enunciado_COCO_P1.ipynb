### Apartado 1: Tetración
La tetración es una hiperoperación que se extiende más allá de la exponenciación. El símbolo empleado para representar esta operación es la doble flecha ascendente en la notación Flecha de Knuth, como se muestra en la siguiente fórmula:

\begin{equation}
a \uparrow\uparrow n = \begin{cases}
a & \text{si } n = 1 \\
a^{(a \uparrow\uparrow (n-1))} & \text{si } n > 1
\end{cases}
\end{equation}

Esto también se puede escribir usando la notación de Rudy Rucker:

\begin{equation}
^{n}a = a^{a^{a^{\cdot^{\cdot^{\cdot^{a}}}}}} \quad \text{(}n \text{ veces)}
\end{equation}

En este apartado, se implementarán dos versiones de un algoritmo para calcular la tetración: una versión recursiva y una versión iterativa.

Tetración Recursiva

```python
""" Parámetros:
    a: int // Es la base de la tetración
    b: int // Es el número de veces por el cual se desea elevar la base por si misma

    Variables:
    resultRec: int // Guarda el valor del resultado dentro de la función recursiva
    resultado: int // Guarda el valor del resultado final y es igual a el resultado de la llamada a la función recursiva

    Funciones:
    tetracionRecursiva: devuelve un int (que no tiene límite como en C) y se encarga de realizar la tetración
    de manera recursiva
"""

def tetracionRecursiva(a,b): #función recursiva que implementa la tetración (similar a las vistas en clase)

    if b == 1:  # Caso específico donde si "b" vale 1 el return es "a" por las propiedades de las potencias
        return a
    
    if b == 0:  # Caso específico donde si "b" vale 0 el return es 1 por las propiedades de las potencias
        return 1
    
    resultRec = a ** tetracionRecursiva(a, b - 1)  #Llamada a la función recursiva. Se eleva b-1 veces como indica el enunciado

    print(f"en el nivel {b}, el resultado es : {resultRec}")

    return resultRec

# Bloque try-except (como try-catch en otros lenguajes) para manejar errores del usuario al ingresar los datos por teclado
try:
    a = int(input("Introduce el número al que le deseas realizar la tetración: ")) # Hacemos uso de la función input() para que el usuario pueda ingresar los datos por teclado. Hay que declarar el tipo de dato de la entrada por teclado por que la función input() devuelve un string
    b = int(input("Introduce el número de veces que deseas elevar ese número por si mismo: "))

    resultado = tetracionRecursiva(a, b)

    print(f"El resultado de realizar la tetración {b} veces a {a} es de: {resultado}")

except ValueError:   # Mensaje de la excepción si el número no es entero
    print("Error al ingresar los datos. Por favor ingrese un número entero.")
```

Tetración iterativa

```python
""" Parámetros:
    a: int // Es la base de la tetración
    b: int // Es el número de veces por el cual se desea elevar la base por si misma

    Variables:
    result: int // Guarda el valor del resultado dentro de la función 
    resultado: int // Guarda el valor del resultado final y es igual a el resultado de la llamada a la función 

    Funciones:
    tetracionIterativa: devuelve un int (que no tiene límite como en C) y se encarga de realizar la tetración
    de manera iterativa 
"""

def tetracionIterativa(a,b):

    if b == 1:
        return a

    if b == 0:
        return 1

    resultItvo = a

    for i in range(1, b):
        resultItvo = a ** resultItvo
        print(f"En la iteración {i}, el resultado parcial es: {resultItvo}")
    
    return resultItvo

try:
    a = int(input("Introduce el número al que le deseas realizar la tetración: ")) # Hacemos uso de la función input() para que el usuario pueda ingresar los datos por teclado. Hay que declarar el tipo de dato de la entrada por teclado por que la función input() devuelve un string
    b = int(input("Introduce el número de veces que deseas elevar ese número por si mismo: "))

    resultado = tetracionIterativa(a, b)
    print(f"El resultado de realizar la tetración {b} veces a {a} es de: {resultado}")

except ValueError:   # Mensaje de la excepción si el número no es entero
    print("Error al ingresar los datos. Por favor ingrese un número entero.")
```

$
S_n(a) = \left\{ a \uparrow\uparrow k \mid k = 1, 2, 3, \ldots, n \right\}
$

Cada término de la sucesión es el resultado de aplicar la operación de tetración $ k $ veces sobre $ \frac{1}{4} $. Es importante considerar que la base es una fracción, lo que puede influir en el comportamiento de la sucesión en comparación con bases enteras o mayores que uno.
